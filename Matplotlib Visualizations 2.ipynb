## B. Plot a pie plot indicating the countries with annual gas prices 

- "less than 1.5" 
- "within 1.5 and 3.0" all inclusive
- "within 3.01 and 4.50" all inclusive
- "within 4.50 and 6.0" all inclusive
- "above 6.0"

## from 1990 to 2008 of each country stated in Question A.  All 4 plots should be in 1 figure. [Legend should be each country]

Solution

```python
#Japan data for ranges

jp = sum(gas_prices['Japan'] < 1.5)
print(jp)
jp1 = sum(gas_prices['Japan'].between(1.5, 3.0 ))# ,inclusive=False)
print(jp1)
jp2 = sum(gas_prices['Japan'].between(3.01, 4.50 ))
print(jp2)
jp3 = sum(gas_prices['Japan'].between(4.50, 6.0 ))
print(jp3)
jp4 = sum(gas_prices['Japan'] > 6.0)
print(jp4)



#South Korea
sk = sum(gas_prices['South Korea'] < 1.5)
print(sk)
sk1 = sum(gas_prices['South Korea'].between(1.5, 3.0 ))# ,inclusive=False)
print(sk1)
sk2 = sum(gas_prices['South Korea'].between(3.01, 4.50 ))
print(sk2)
sk3 = sum(gas_prices['South Korea'].between(4.50, 6.0 ))
print(sk3)
sk4 = sum(gas_prices['South Korea']> 6.0)
print(sk4)
'/n'
#UK
uk = sum(gas_prices['UK'] < 1.5)
print(uk)
uk1 = sum(gas_prices['UK'].between(1.5, 3.0 ))# ,inclusive=False)
print(uk1)
uk2 = sum(gas_prices['UK'].between(3.01, 4.50 ))
print(uk2)
uk3 = sum(gas_prices['UK'].between(4.50, 6.0 ))
print(uk3)
uk4 = sum(gas_prices['UK'] > 6.0)
print(uk4)


#Germany
ger = sum(gas_prices['Germany'] < 1.5)
print(ger)
ger1 = sum(gas_prices['Germany'].between(1.5, 3.0 ))# ,inclusive=False)
print(ger1)
ger2 = sum(gas_prices['Germany'].between(3.01, 4.50 ))
print(ger2)
ger3 = sum(gas_prices['Germany'].between(4.50, 6.0 ))
print(ger3)
ger4 = sum(gas_prices['Germany']> 6.0)
print(ger4)
```

```python
#plt.suptitle('World Gas Prices from 1990-2008')
#Ploting our pieplots

plt.figure(figsize=(12,12))
labels = ['<1.5','1.5-3.0','3.01-4.50','4.50-6.0','>6.0']#['Japan','Japan','Japan','Japan','Japan']
colors = ['pink','teal','violet','black','navy']

#Japan
plt.subplot(2,2,1)
explode = [0,0.5,0,0,0]
plt.title('World gas prices for Japan 1990-2008')
plt.pie([jp,jp1,jp2,jp3,jp4],labels=labels,colors=colors,autopct='%.3f%%',explode=explode)
plt.legend()

#South Korea
plt.subplot(2,2,2)
explode = [0,0,0.3,0,0]
plt.title('World gas prices for South Korea 1990-2008')
plt.pie([sk,sk1,sk2,sk3,sk4],labels=labels,colors=colors,autopct='%.3f%%',explode=explode)
plt.legend()

#UK
plt.subplot(2,2,3)
explode = [0,0,0.3,0,0]
plt.title('World gas prices for UK 1990-2008')
plt.pie([uk,uk1,uk2,uk3,uk4],labels=labels,colors=colors,autopct='%.3f%%',explode=explode)
plt.legend()

#Germany
plt.subplot(2,2,4)
explode = [0,0,0.3,0,0]
plt.title('World gas prices for Germany 1990-2008')
plt.pie([ger,ger1,ger2,ger3,ger4],labels=labels,colors=colors,autopct='%.3f%%',explode=explode)
plt.legend()


plt.tight_layout()
plt.show()
```
