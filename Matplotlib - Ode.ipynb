## Subplot


```python
#plt.subplot(a,b,c) a is for the number of rows and b is for column and c is for position in your plot table
#plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
#plt.subplot(3,1,3)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Plot 1')

#plot 2
a = np.array([0,1,2,3])
b = np.array([10,20,30,40])
plt.subplot(2,2,2)
plt.plot(a,b)
plt.title('Plot 2')

#plot 3
c = np.array([0,1,2,3])
d = np.array([10,20,30,40])
plt.subplot(2,2,3)
plt.plot(c,d)
plt.title('Plot 3')

plt.tight_layout()
plt.show()
```

## Scatter Plots

```python
#x = np.linspace(a,b,c)a is the starting point, b is your interval space and c is the number of data points you want on your graph
#plt.grid() is to put a grid on your plot

x = np.linspace(0,10,30)
y = np.sin(x)
plt.scatter(x,y, marker = '*', color = 'b')
plt.title('Our sine Graph')
plt.xlabel('x axis')
plt.ylabel;('sinX')
plt.grid()
plt.show()
```

## Grouped Bar Chart

```python
plt.figure(figsize = (5,5))

labels=['Mon','Tue','Wed','Thu','Fri']
mobile_sales = [20,34,30,35,27]
pc_sales = [25,32,34,20,25]

x = np.arange(len(labels))
width = 0.35
a = plt.bar(x-width/2,mobile_sales,width,color='b',label='Mobile')
a = plt.bar(x+width/2,pc_sales,width,color='violet',label='PC')

plt.ylabel('Total Sales')
plt.xlabel('Total sales by moblie and pc')
plt.xticks(x,labels) #gets location for the labels on x-axis
plt.legend()
#plt.bar_label(a) #bar label adds labels to bars
#plt.bar_label(b)
#plt.savefig('groupedbar1.png')
plt.show()
```

## Stacked Bar Chart

```python
#datat for 2016 olympics

plt.figure(figsize=(5,7))

x = ['USA','GBR','China','Russia','Germany']
gold = [46,27,26,19,17]
silver = [37,23,18,18,10]
w=0.5

plt.bar(x,gold,w, color ='k')
plt.bar(x,silver,w, bottom=gold,color='pink')

plt.show()
```
