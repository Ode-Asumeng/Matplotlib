### Scatter plot activity
Create a scatter plot with matplotlib that displays xt paired with yt and zt:

```python
xt=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt
```

Change the shape of the points to

<br> Also:

#### Solution for Scatter Plot Activity

```python
plt.figure(figsize=(5,5))

xt=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt

#plt.scatter(xt, marker = '*', color = 'indigo', label='o')
plt.scatter(xt,yt, marker = 's', color = 'black', label='s')
plt.scatter(xt,zt, marker = '^', color = 'green',  label='^')
plt.title('Scatter Plot with 3 variables')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.grid()
plt.show()
```

### Grouped Bar plots
create a grouped bar plot of the data below:
<br> Indicate the ff
- label
- title
- bar_label
- xticks
- legend

```python
#data
countries=['Ghana','Portugal','South Korea','Uruguay','Germany']
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,30,25,18,24]
```

#### Solution of Grouped Bar plots

```python
plt.figure(figsize = (5,5))

countries=['Ghana','Portugal','South Korea','Uruguay','Germany']
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,30,25,18,24]
width = 0.2
x = np.arange(len(countries))

plt.bar(x-0.2,gold,width,color='magenta',label='Gold')
plt.bar(x,silver,width,color='grey',label='Silver')
plt.bar(x+0.2,bronze,width,color='black',label='Bronze')
plt.xticks(x,['Ghana','Portugal','South Korea','Uruguay','Germany'])
plt.xlabel('Participating Countries')
plt.ylabel('Number of Trophies won')
plt.legend()
plt.show()
```

## What are the insights you made?
Write them in this cell

- Ghana had the highest amount of awards overall.
- Germany had the least amount of silver medals out of the 5 countries.
- Uruguay had the lowest amount of bronze medals out of the 5 countries.
- Portugal had the second highest amount of awards overall.
